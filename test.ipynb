```python
classifier = Agent(
    role = "email classifier",
    goal = "classify email as spam or not spam",
    backstory="I am a machine learning model that can classify emails as spam or not spam",
    allow_delegation = False,
    llm = model
)
```

```python
responder = Agent(
    role = "email responder",
    goal = "respond to email",
    backstory="I am a human that can respond to emails",
    allow_delegation = False,
    llm = model
)
```

```python
classify_email = Task(
    description = "classify email as spam or not spam",
    agent = classifier,
    expected_output = "spam or not spam"
)
```

```python
respond_to_email = Task(
    description = "respond to email",
    agent = responder,
    expected_output = "very consize response"
)
```

```python
crew = Crew(
    agents=[classifier, responder],
    tasks=[classify_email, respond_to_email],
    verbose=2,
    process=Process.sequential
)
```

```python
output = crew.kickoff()
```
