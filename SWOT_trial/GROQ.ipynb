```python
data_analyst = Agent(
    role="Data Analyst",
    goal="Get the data of the student and generate meaningful insights",
    backstory="I am a data analyst and I have been working in this field for the past 5 years. I have worked on various projects and have experience in data analysis. I have a good understanding of data and can generate meaningful insights from it.",
    allow_delegation=False,
)
```

```python
insights_generator = Task(
    description=f"Generate meaningful insights from the data of the student. without using numerical values just take the insights from it. if he/her got zero in any subject just do not include that subject in the insights.include the suject insights which got low marks. use this infirmation to generate insights: {student_info(100)} do not use direct data in response. do not use words like score marks etc.",
    agent= data_analyst,
    expected_output = "The insights generated from the data of the student",
)
```

```python
SWOTER = Agent(
    role = "Ambious Person",
    goal = "Create a SWOT analysis for future carrer",
    backstory="""To conduct a personal SWOT analysis for future career planning, start by dividing a sheet of paper or a digital document into four quadrants labeled Strengths, Weaknesses, Opportunities, and Threats. Begin by listing your strengths, considering your natural talents, skills, achievements, and resources. Reflect on questions such as what you are naturally good at, what advantages you have over others, and what achievements set you apart from your colleagues. Next, identify your weaknesses by acknowledging areas where you need improvement, bad habits, and skills that require development. Ask yourself about your negative traits, areas where your education or training may be lacking, and situations where you lack confidence. Then, explore opportunities for growth within your industry, company, and desired career path by considering the state of your industry, emerging technologies, and the potential support from your professional network. Finally, address potential threats to your career goals, such as external factors like industry changes, economic conditions, and competition, as well as internal challenges like bad habits and skill deficiencies. Use this SWOT analysis to identify goals, devise strategies, and create action plans for advancing your career, ensuring an honest and balanced assessment to stay motivated and effective. An example of a personal SWOT analysis might include strengths like being a creative thinker, working well under pressure, having an advanced degree, being organized and personable, being resourceful, and asking good questions. Weaknesses might include procrastination, tardiness, past issues with coworkers, less experience with AI technology, shyness in meetings, and presentation anxiety. Opportunities might involve proposing new market strategies, utilizing AI tools, seeking advice from mentors, taking online courses, and working with a career coach. Threats could include the need for more education, slower project completion compared to peers, being viewed as unreliable due to bad habits, potential job obsolescence due to AI advancements, and increasing competition from new talent. Use this SWOT analysis to brainstorm goals and strategize action plans for career advancement, ensuring an honest and balanced approach to create an effective and motivating assessment.""",
    allow_delegation = False,
)
```

```python
FORMATTER = Agent(
    role = "chatbot response formatter",
    goal = "you have to carefully format the response to the four quadrants of SWOT analysis",
    backstory="You are very good at formatting the text data and you have to format the response to the four quadrants of SWOT analysis. try make response more like human.",
    allow_delegation = False,
)
```

```python
SWOT_analysis = Task(
    description = f"""Create a SWOT analysis for a student using the provided detailed information. Focus on uncovering valuable insights that highlight the student's strengths and areas for growth beyond academic scores. For example, instead of focusing on specific scores, consider their skills, interests, and experiences that contribute positively to their academic journey. Identify strengths that set them apart, areas where improvement could enhance their abilities, opportunities for future development based on their interests, and potential challenges they may encounter. Ensure the analysis is constructive and balanced, emphasizing their potential and positive attributes. Use supportive and motivating language to inspire confidence in the student about their capabilities and future opportunities""",
    agent = SWOTER,
    expected_output = "Perfect SWOT analysis for the student.",
)
```

```python
formating = Task(
    description = """take the SWOT analysis and format it for human readable format.remove unnecessary special characters like *. use '-' for bullet points.remove the mini title from in front of every point. do not add any extra title or footer line just four quadrants with their respective points.""",
    agent = FORMATTER,
    expected_output = "python dictionary with formatted SWOT analysis",
)
```

```python
crew = Crew(
    agents=[data_analyst,SWOTER, FORMATTER],
    tasks=[insights_generator,SWOT_analysis, formating],
    verbose=2,
    process=Process.sequential
)
```

```python
output = crew.kickoff()
```
